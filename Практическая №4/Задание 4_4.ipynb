```python
import numpy
import pandas as pd
import matplotlib.pyplot as mtp
df = pd.read_csv("winequality-white.csv", ";")
df = df.loc[df["residual sugar"] < 1]
df = df[["density", "pH"]]
df = df.iloc[df.index % 2 == 0]
df["pH"].hist()
mtp.title("pH")
mtp.show()
df["density"].plot()
mtp.title("density")
mtp.show()
```
